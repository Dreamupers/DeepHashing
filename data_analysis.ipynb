```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import random
```

### Flickr25k

```python
df = pd.read_table("./data/flickr25k/allannots.txt", sep=" ", header=None)
lables = np.array(df.values[:, 1:])
print(lables.shape[0])
lables = lables.sum(axis=0)
print(lables.min(), np.median(lables), lables.max())
class_name = ['animals', 'baby', 'bird', 'car', 'clouds',
'dog', 'female', 'flower', 'food', 'indoor',
'lake', 'male', 'night', 'people', 'plant_life',
'portrait', 'river', 'sea', 'sky', 'structures',
'sunset', 'transport', 'tree', 'water']

plt.figure(figsize=(10, 3))
ax = sns.barplot(x=class_name, y=lables)
ax.bar_label(ax.containers[0])
plt.xticks(rotation="vertical")
plt.yscale("log")
plt.yticks([100,1000,10000])
plt.show()
```
